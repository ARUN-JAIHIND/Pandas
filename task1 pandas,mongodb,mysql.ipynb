```python
#Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset
#   
```

```python
df['working']=(df['ActivityDate']==df['ActivityDate'])
```

```python
def cal(a):
    if a==True:
        return 1
    else:
        return 0
```

```python
df['working']=df['working'].apply(cal)
```

```python
meter=df.groupby('Id')[['working','Calories']].sum()
```

```python
meter['avg_calburn']=meter['Calories']/meter['working']
```

```python
def check(a):
    if a>1500:
        return 1
    else:
        return 0
```

```python
meter['Health']=meter['avg_calburn'].apply(check)
```

```python
#Total Healthy person
meter['Health'].sum()
```

```python
#Assume regular person activity greater than equal to 15 work day
def reg(a):
    if a>=15:
        return 'regular'
    else:
        return 'unregular'
```

```python
meter['status']=meter['working'].apply(reg)
```
