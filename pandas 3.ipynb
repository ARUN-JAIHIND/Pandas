```python
data4 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
```

```python
data5  = {'emp_id' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
df4
```

```python
df5
```

```python
#for merge by default inner join used i.e. only inersection data shows w.r.t. same column name in both data frame(emp_id)
pd.merge(df4,df5)
```

```python
#merge left join 
pd.merge(df4,df5 ,  how ='left')
```

```python
#merge right join
pd.merge(df4,df5 ,  how ='right')
```

```python
# on : Column or index level names to join on. 
#These must be found in both DataFrames. 
#If on is None and not merging on indexes then this defaults to the intersection of the columns in both DataFrames
pd.merge(df4, df5  , how = 'right' , on = 'emp_id')
```

```python
data6 = {'emp_id1' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
         
data7  = {'emp_id2' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
#left_on :Column or index level names to join on in the left DataFrame
#right_on :Column or index level names to join on in the right DataFrame
#if column name is not same  but if we want to merge the we can use left_on,right_on
pd.merge(df6,df7 , left_on='emp_id1' , right_on='emp_id2',how = 'inner')
```

```python
df_sales=pd.read_csv(r"C:\Users\MIET\Downloads\fsds\sales_data_final.csv")
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if a>=0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(50)
```

```python
profit_flag(-40)
```

```python
df_sales['flag_profit']=df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
def ship(a):
    if a>35:
        return 's_cost high'
    else:
        return 's_cost low'
```

```python
df_sales['shipping']=df_sales['shipping_cost'].apply(ship)
```

```python
df_sales.head()
```

```python
#we can apply inbuild function also in dataframe
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a) : 
    if  a < 10 :
        return 'low'
    elif a >10 and a <20 :
        return 'medium'
    else :
        return 'high'
```

```python
df_sales['flag_quantity']= df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
```
